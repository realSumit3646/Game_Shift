```python
from ultralytics import YOLO
import cv2

# Load the YOLO model
model = YOLO(r"C:\Users\stude\Downloads\YOLO\runs\detect\train2\weights\best.pt")

# Open the webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Perform detection
    results = model(frame)

    # Visualize the results on the frame
    annotated_frame = results[0].plot()  # Annotates the frame with bounding boxes and labels

    # Display the frame
    cv2.imshow("YOLO Detection", annotated_frame)

    # Break the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the OpenCV window
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO

# Load the YOLO model
model = YOLO("yolo11n.pt")

# Paths to dataset configuration files
dataset1 = r"C:\Users\stude\Downloads\YOLO\yawning\data.yaml"  # Dataset 1
dataset2 = r"C:\Users\stude\Downloads\YOLO\eye_mouth_average_dataset\data.yaml"  # Dataset 2

# Parameters
total_epochs = 20
img_size = 640
device = 0  # GPU device
epochs_per_dataset = 1  # Number of epochs per dataset before switching

# Custom Training Loop
for epoch in range(0, total_epochs, epochs_per_dataset * 2):
    # Train on Dataset 1
    print(f"Training on Dataset 1 (Epochs {epoch + 1}-{epoch + epochs_per_dataset})")
    model.train(
        data=dataset1,
        epochs=epochs_per_dataset,
        imgsz=img_size,
        device=device,
        resume=False  # Prevents overwriting model history
    )

    # Train on Dataset 2
    print(f"Training on Dataset 2 (Epochs {epoch + epochs_per_dataset + 1}-{epoch + 2 * epochs_per_dataset})")
    model.train(
        data=dataset2,
        epochs=epochs_per_dataset,
        imgsz=img_size,
        device=device,
        resume=False
    )
```
